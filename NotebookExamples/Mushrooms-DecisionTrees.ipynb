# PoisonousMushrooms

"This data set includes descriptions of hypothetical samples corresponding to 23 species of gilled mushrooms in the Agaricus and Lepiota Family (pp. 500-525). Each species is identified as definitely edible, definitely poisonous, or of unknown edibility and not recommended. This latter class was combined with the poisonous one. The Guide clearly states that there is no simple rule for determining the edibility of a mushroom; no rule like ``leaflets three, let it be'' for Poisonous Oak and Ivy."

_ source: http://archive.ics.uci.edu/ml/datasets/Mushroom _

The objective will be to create a rule based model combining different decisions to classify a mushroom on edible or poisonous. A Decision Tree classifier will be used to extract the rules.

Every row on the file looks like this:
> p,x,s,n,t,p,f,c,n,k,e,e,s,s,w,w,p,w,o,p,k,s,u

The first character is the status: poisonous(p) / edible(e), and the rest are the features, all of them categorical

#### Attribute Information:

1. cap-shape: bell=b,conical=c,convex=x,flat=f, knobbed=k,sunken=s 
2. cap-surface: fibrous=f,grooves=g,scaly=y,smooth=s 
3. cap-color: brown=n,buff=b,cinnamon=c,gray=g,green=r, pink=p,purple=u,red=e,white=w,yellow=y 
4. bruises?: bruises=t,no=f 
5. odor: almond=a,anise=l,creosote=c,fishy=y,foul=f, musty=m,none=n,pungent=p,spicy=s 
6. gill-attachment: attached=a,descending=d,free=f,notched=n 
7. gill-spacing: close=c,crowded=w,distant=d 
8. gill-size: broad=b,narrow=n 
9. gill-color: black=k,brown=n,buff=b,chocolate=h,gray=g, green=r,orange=o,pink=p,purple=u,red=e, white=w,yellow=y 
10. stalk-shape: enlarging=e,tapering=t 
11. stalk-root: bulbous=b,club=c,cup=u,equal=e, rhizomorphs=z,rooted=r,missing=? 
12. stalk-surface-above-ring: fibrous=f,scaly=y,silky=k,smooth=s 
13. stalk-surface-below-ring: fibrous=f,scaly=y,silky=k,smooth=s 
14. stalk-color-above-ring: brown=n,buff=b,cinnamon=c,gray=g,orange=o, pink=p,red=e,white=w,yellow=y 
15. stalk-color-below-ring: brown=n,buff=b,cinnamon=c,gray=g,orange=o, pink=p,red=e,white=w,yellow=y 
16. veil-type: partial=p,universal=u 
17. veil-color: brown=n,orange=o,white=w,yellow=y 
18. ring-number: none=n,one=o,two=t 
19. ring-type: cobwebby=c,evanescent=e,flaring=f,large=l, none=n,pendant=p,sheathing=s,zone=z 
20. spore-print-color: black=k,brown=n,buff=b,chocolate=h,green=r, orange=o,purple=u,white=w,yellow=y 
21. population: abundant=a,clustered=c,numerous=n, scattered=s,several=v,solitary=y 
22. habitat: grasses=g,leaves=l,meadows=m,paths=p, urban=u,waste=w,woods=d

Some conclusions from the visualization:

    - there are a few obvious rules than can be found, where the is a big difference between poisonous and edible on some classes, like the mushrooms is poisonous if:

        - if gill-color is b(buff)
        - if population is v(several)
        and so on

The idea is for a C4.5 decision tree to find all or most of these rules and combinations

```python
edibility = md["edibility"]
```

```python
del(md["edibility"])
```

Creation of dummy variables, to enconde the categorical variables:

```python
md_encoded = pandas.get_dummies(md)
```

```python
md_encoded.tail()
```

Creation of the testing set, to check the trained model on different data:

```python
import sklearn.cross_validation
md_train, md_test, ed_train, ed_test = sklearn.cross_validation.train_test_split(md_encoded, edibility, test_size = 0.333)
```

The tree:

```python
from sklearn import tree
md_tree = tree.DecisionTreeClassifier()
```

```python
md_tree.fit(md_train, ed_train)
```

Testing the tree:

```python
ed_prediction = md_tree.predict(md_test)
```

```python
from sklearn import metrics
score = metrics.accuracy_score(ed_test, ed_prediction)
```

```python
score
```

A perfect score! So it is possible to create a decision tree to perfectly classify mushrooms in edible and poisonous based on physical characteristics

It is possible to visualize the tree, following the instructions on the scikit-learn Decision Trees page

```python
from sklearn.externals.six import StringIO  
from IPython.display import Image  
```

```python
dot_data = StringIO()  
tree.export_graphviz(md_tree, out_file=dot_data, feature_names=md_encoded.columns, class_names=list(md_tree.classes_), 
                     filled=True, rounded=True, impurity=False, special_characters=True)
```

```python
#It is necessary to install GraphViz and pydot-ng first
import pydot_ng as pydot
graph = pydot.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())  
```

In this tree blue labels are poisonous and red are edible, and <= 0.5 indicates absence of the especific class.

This representation could be used by a human on the field to decide if some mushroom is edible or poisonous, although given the risk it is not recommended to actualy eat it if the user don't have experience with mushrooms
